```python
sf['Date']=pd.to_datetime(sf['Date'])
sf['Month']=sf.Date.dt.month_name().str[:3]
sf['Year']=sf.Date.dt.year
sf['Weekday']=sf.Date.dt.weekday
sf.Weekday=sf.Weekday.map({0: "Monday",1: "Tuesday",2: "Wednesday" ,3: "Thursday",4:"Friday",5:"Saturday",6:"Sunday"})
sf["Weekday"] = pd.Categorical(sf["Weekday"], categories=
    ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday', 'Sunday'],
    ordered=True)
sf.head()
```

```python
sf.info()
```

```python
fig,((ax1,ax2),(ax3,ax4)) = mtb.subplots(2,2,figsize=(16,8), gridspec_kw={'hspace': 0.3})

sns.barplot(data=sf,x="Seasons",y="Rented Bike Count",palette="colorblind",ax=ax1,ci=None)
# sns.barplot(data=sf,x="Wind speed (m/s)",y="Rented Bike Count",palette="colorblind",ax=ax1,ci=None)
sns.lineplot(data=sf,x="Month",y="Rented Bike Count",palette="colorblind",ax=ax3,ci=None)
sns.barplot(data=sf,x="Weekday",y="Rented Bike Count",palette="colorblind",ax=ax2,ci=None)
sns.lineplot(data=sf,x="Hour",y="Rented Bike Count",color="tab:blue",ax=ax4,ci=None)
ax1.set_title("Seasonal Bike Rents")
ax3.set_title("Monthly Bike Rents")
ax1.set_xlabel("")
ax3.set_xlabel("")
ax2.set_title("Weekday Bike Rents")
ax4.set_title("Hourly Bike Rents")
ax2.set_xlabel("")
mtb.savefig("Timeline.png",dpi=300)

mtb.show()
```

***Seasonal Bike Rents***
**Lets take a look at the first chart. From the given bar plot we can clearly see that the rented bike count is less in winters being highest in the summer followed by Autumn and Spring. From the graph we can make an assumption that people like to rent bikes when temperature is higher, but we can also say that in seoul many students have summer vacations, so they use rented bikes to visit places or commute to their summer camps**

***Weekly Bike Rents***
**Now lets move over to the Weekly graph. From the graph it is clearly visible that there is a rise in the number of the rented bikes and it declines on Sunday which is day off to maximum students and employees**

***Monthly Bike rents***
From this graph it is clearly visisble that there Is slow rise in the count of rented bikes from the months January to Febuary and it begins to rise from february and peeks at the starting weeks of June and begins declinig From june till September. This is probably due to the start of the rainy season, it is hard and less safe to ride bikes in the rainy season so most of them prefer using a public transport or a car or prefer commuting to their respective destination with an umbrella. ***Hourly Bike Rents***
By taking a look at this graph it can be clearly seen that there is a sharp rise in the mornung hours that is between 6 to 10 and evening hours 4pm to 7 pm. It can be deducted from this graph that the maximum rented bikes are used by students and employees who commute from home to their respective destinations in the morning and vice versa in the evening.

```python
u_var=sf.iloc[:,1:-3].corr()
sns.heatmap(data=u_var)
```

From the above heatmap we can see the corelation between the Rented bike count and the other Whether factors affecting the bike count. It is clear from the plot that the Bike count is majorly affected by the Temperature, Humidity,Visiblity and Solar Radiation.

```python
sns.scatterplot(y='Rented_Bike_Count',x='Temperature',data=sf)
```

The rental bike count decreases when the temperature drops below 0. People avoid to rent bikes on lower temperatures that is the count is usally less in the winter months, ans most of them prefer public or private transports.

```python
sns.scatterplot(x='Rented_Bike_Count',y='Dew_point_temperature',data=sf)
```

The rental bike count is less for lower dew point temperatures.
