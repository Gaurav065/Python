```python
sf['Date']=pd.to_datetime(sf['Date'])
sf['Month']=sf.Date.dt.month_name().str[:3]
sf['Year']=sf.Date.dt.year
sf['Weekday']=sf.Date.dt.weekday
sf.Weekday=sf.Weekday.map({0: "Monday",1: "Tuesday",2: "Wednesday" ,3: "Thursday",4:"Friday",5:"Saturday",6:"Sunday"})
sf["Weekday"] = pd.Categorical(sf["Weekday"], categories=
    ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday', 'Sunday'],
    ordered=True)
sf.head()
```

```python
fig,((ax1,ax2),(ax3,ax4)) = mtb.subplots(2,2,figsize=(16,8), gridspec_kw={'hspace': 0.3})

sns.barplot(data=sf,x="Seasons",y="Rented Bike Count",palette="colorblind",ax=ax1,ci=None)
# sns.barplot(data=sf,x="Wind speed (m/s)",y="Rented Bike Count",palette="colorblind",ax=ax1,ci=None)
sns.lineplot(data=sf,x="Month",y="Rented Bike Count",palette="colorblind",ax=ax3,ci=None)
sns.barplot(data=sf,x="Weekday",y="Rented Bike Count",palette="colorblind",ax=ax2,ci=None)
sns.lineplot(data=sf,x="Hour",y="Rented Bike Count",color="tab:blue",ax=ax4,ci=None)
ax1.set_title("Seasonal Bike Rents")
ax3.set_title("Monthly Bike Rents")
ax1.set_xlabel("")
ax3.set_xlabel("")
ax2.set_title("Weekday Bike Rents")
ax4.set_title("Hourly Bike Rents")
ax2.set_xlabel("")
mtb.savefig("Timeline.png",dpi=300)

mtb.show()
```

***Seasonal Bike Rents***
**Lets take a look at the first chart. From the given bar plot we can clearly see that the rented bike count is less in winters being highest in the summer followed by Autumn and Spring. From the graph we can make an assumption that people like to rent bikes when temperature is higher**

***Weekly Bike Rents***
**Now lets move over to the Weekly graph. From the graph it is clearly visible that there is a rise in the number of the rented bikes and it declines on Sunday which is day off to maximum students and employees**

***Monthly Bike rents***
From this graph it is clearly visisble that there Is slow rise in the count of rented bikes from the months January to Febuary and it begins to rise from february and peeks at the starting weeks of June and begins declinig From june till September. This is probably due to the start of the rainy season, it is hard and less safe to ride bikes in the rainy season so most of them prefer using a public transport or a car or prefer commuting to their respective destination with an umbrella. The count begins to rise as the rainy seasons ends and the fall/autumn begins to start as it is the time of the year marked with clear skys. And after seeing the peek in October it begans to fall down as the temperature begins to fall and Winter approaches.


***Hourly Bike Rents***
By taking a look at this graph it can be clearly seen that there is a sharp rise in the mornung hours that is between 6 to 10 and evening hours 4pm to 7 pm. It can be deducted from this graph that the maximum rented bikes are used by students and employees who commute from home to their respective destinations in the morning and vice versa in the evening.
